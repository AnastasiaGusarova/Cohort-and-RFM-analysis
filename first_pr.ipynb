## 5. Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц. 

Retention rate показывает отношение клиентов, которые вернулись за повторными покупками, к числу клиентов на начало расчетного периода.

Определила четыре параметра для проведения когортного анализа:

- Признак формирования когорты — первая покупка
- Размер когорты — временной интервал: месяц.
- Отчетный период — 2017 год январь - декабрь.
- Анализируемый ключевой показатель: Retention Rate

```python
# Скопировала датасет в новую переменную
for_cohort = task.query('order_status not in @status').copy()
```

```python
# Проверка на полный год
for_cohort.purchase_month.sort_values().unique()
```

```python
# Добавила поле для фильтрации по 2017 году
for_cohort['purchase_year'] = for_cohort.order_purchase_timestamp.dt.year
```

```python
# Отфильтровала по 2017 году
for_cohort2 = for_cohort.query('purchase_year == 2017')
```

```python
for_cohort2.customer_unique_id.nunique()
```

Чтобы построить когорты, нам нужно сгруппировать клиентов по дате их первой покупки. Такая информация указана не была, поэтому создаем новый датасет с данными о первой покупки клиента. 

```python
first_orders = for_cohort2.groupby('customer_unique_id', as_index = False)\
                          .agg({'purchase_month': 'min'})\
                          .rename(columns={"purchase_month": "first_purchase"})
first_orders
```

```python
first_orders.dtypes
```

Дальше добавим столбец с месяцем первой покупки для построения когорт

```python
# Добавим дату первой покупки с помощью метода merge() и сохраним получившийся датафрейм в переменную orders_merged
orders_merged = for_cohort2.merge(first_orders, how='left', on='customer_unique_id')\
                           .reset_index(drop=True)
orders_merged
```

```python
# Группируем по месяцу покупки, по дате первой покупки и посчитываем количество уникальных клиентов  и количество покупок
cohorts = orders_merged.groupby(['first_purchase', 'purchase_month'])['customer_unique_id'] \
                                .agg(['nunique', 'count']) \
                                .rename(columns={'nunique': 'unique_clients', 'count': 'count_orders'})
```

#### Когорты готовы

Добавим порядковое значение CohortPeriod для каждого из OrderPeriod. CohortPeriod определим как порядковый номер месяца в массиве начинающийся с 1

```python
def cohort_period(df):
    df['CohortPeriod'] = np.arange(len(df)) + 1 # отсчет с 1
    return df
cohorts = cohorts.groupby(level=0).apply(cohort_period)
cohorts
```

Делаем сводную таблицу, так чтобы в строке была информация о когорте, а в столбце значение за период.

```python
cohort_pivot = cohorts.pivot_table(index = 'first_purchase',
                                     columns = 'CohortPeriod',
                                     values = 'unique_clients')
cohort_pivot
```

Чтобы получить retantion rate (retention_matrix) делим каждое значение строк на первое значение с строке, которое является размером когорты (cohort_size).

```python
cohort_size = cohort_pivot.iloc[:,0]
retention_matrix = cohort_pivot.divide(cohort_size, axis = 0)
retention_matrix
```

#### Вывод задание 5:
Когорта с самым высоким retention на 3й месяц за 2017 год у покупателей, которые совершили первую покупку 2017-05. Retention rate составляет 0.48%

```python
#вариант визуализации, heatmap
sns.set(style='ticks')
plt.figure(figsize=(24, 16))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_matrix, mask=retention_matrix.isnull(), annot=True, fmt='.2%')
```

## 6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. 

В кластеризации можешь выбрать следующие метрики: 
- R - время от последней покупки пользователя до текущей даты, 
- F - суммарное количество покупок у пользователя за всё время, 
- M - сумма покупок за всё время. 

Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. 

#### В данном случае я решила использовать все данные, за все время, так как покупателей более чем с одной покупкой очень мало. И кроме того, покупатели у которых более одной покупки сделали это в то же время что и первую покупку. 

#### Поэтому использование frequency не целесоообразно с этими данными, но в рамках задания будет присутствовать. 


```python
# скоприровала df
for_RFM = task.copy()
```

```python
# фильтрация по статусам
for_RFM = for_RFM.query('order_status not in @status')
```

```python
# оставила только нужные для RFM анализа столбцы
for_RFM = for_RFM[['customer_id','order_purchase_timestamp','order_id','price']]
```

```python
for_RFM.sort_values('order_purchase_timestamp', ascending = False)
```

#### Последний заказ в таблице был сделан 2018-09-03 09:06:57. Данную дату буду использовать для расчета давности для функции NOW. 

R - время от последней покупки пользователя до текущей даты: NOW - order_purchase_timestamp.max

F - суммарное количество покупок у пользователя за всё время: буду использовать order_id для подсчета, 

M - сумма покупок за всё время: price.sum

И все это в группировке по customer_id

```python
NOW = pd.to_datetime('2018-09-03 09:06:57', format = '%Y-%m-%d %H:%M:%S')
NOW
```

```python
RFM_table=for_RFM.groupby('customer_id', as_index = False).agg({'order_purchase_timestamp': lambda x: (NOW - x.max()).days, # Recency  время от последней покупки пользователя до текущей даты
                                                'order_id': 'count', # Frequency
                                                'price': 'sum'})    # Monetary 
```

```python
RFM_table.dtypes
```

```python
RFM_table.rename(columns={'order_purchase_timestamp': 'recency', 
                         'order_id': 'frequency',
                         'price': 'monetary'}, inplace=True)
RFM_table
# переименовала столбцы
```

#### Собрала таблицу RFM. Получила значения для каждого покупателя. 

```python
describe = RFM_table.describe().transpose()
describe
```

#### Теперь можно сегментировать покупателей в зависимости от давности покупки, частоты и суммы. Для разбивки на кластеры буду использовать квантили, кроме frequency.  Квантили в таблице describe. Recency и Monetary поделю на 4 кластера.

Использовала обычную группировку по покупателю для их подсчета и фильтрацию по количеству покупок. 
Так как у большенства покупателей меньше 2 покупок, а больше 3 покупок есть только у 965 покупателей, поделю Frequency на 3 кластера: 0-1; 2-3, 4-21, границы включительно.

#### Ранги будут ставиться по следующиму принципу. 

В разных источниках ранги строятся по разному, в одних 4 это лучший ранг, в других это 1. Я выбрала 1 за лучший ранг. Таким образом, 111 будут лучшие клиенты, 434 - худшие.  Получается что чем больше ранг тем хуже клиент.

По давности заказа (Recency) по квантилям:

 - 1 — недавние клиенты x <= 0.25;
 - 2 - x <= 0.50
 - 3 - x <= 0.75
 - 4 — давние клиенты else;

По сумме покупок (Monetary) по квантилям:

 - 4 — маленькая сумма x <= 0.25.
 - 3 - x <= 0.50
 - 2 - x <= 0.75 
 - 1 — большая сумма else;

По частоте покупок (Frequency):

 - 3 — единичные покупки x <= 1;
 - 2 —  x <= 3 ;
 - 1 — покупают часто else.
 

```python
quantiles = RFM_table.quantile(q=[0.25,0.5,0.75])
quantiles = quantiles.to_dict()
# добавлю табличку  и переведу в словарь для удобства
```

```python
quantiles
```

```python
def r_score(x):
    if x <= quantiles['recency'][0.25]:
        return 1
    elif x <= quantiles['recency'][0.50]:
        return 2
    elif x <= quantiles['recency'][0.75]: 
        return 3
    else:
        return 4
# функции для применения к столбцу recency в таблице
```

```python
def m_score(x):
    if x <= quantiles['monetary'][0.25]:
        return 4
    elif x <= quantiles['monetary'][0.50]:
        return 3
    elif x <= quantiles['monetary'][0.75]: 
        return 2
    else:
        return 1
```

А для frequency функция будет немного другой. 

```python
def f_score(x):
    if x <= 1:
        return 3 
    elif x <= 3:
        return 2
    else:
        return 1
```

```python
RFM_table['R'] = RFM_table['recency'].apply(lambda x: r_score(x))
RFM_table['F'] = RFM_table['frequency'].apply(lambda x: f_score(x))
RFM_table['M'] = RFM_table['monetary'].apply(lambda x: m_score(x))

# применяем созданные функции к таблице и создаем 3 столбца с ранговыми значениями. 
```

```python
RFM_table
```

```python
RFM_table['RFM Score'] = RFM_table['R'].map(str) + RFM_table['F'].map(str) + RFM_table['M'].map(str)

# функция map() используется для применения функции к каждому элементу итерируемого объекта  
# и возврата нового итератора для получения результатов. 
# Применяем чтобы перевести элементы столбцов R F M в строку и сконкатенировать  для столбца RFM Score 
```
